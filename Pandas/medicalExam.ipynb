# Pandas - Medical Exam

```python
import pandas as pd
df = pd.read_csv('medical_exam.csv')
df.head()
```

Agregar la columna 'overweight': 

```python
df['overweight'] = None
df.head()
```

```python
df1 = df[(df['cholesterol'] <= 1) & (df['gluc'] <= 1)]
df1['overweight'] = 0
df1.shape
```

```python
df2 = df[(df['cholesterol'] > 1) | (df['gluc'] > 1)]
df2.overweight = 1
df2.shape
```

```python
print(df1.shape[0] + df2.shape[0])
```

```python
df3 = pd.concat([df1, df2])
df3.shape
```

```python
df4 = df3[['cholesterol', 'gluc', 'smoke', 'alco', 'active', 'cardio', 'overweight']]
df4 = pd.melt(df4)
df4 
```

```python
df5 = df4[df4['variable'] == 'cardio']
df5
```

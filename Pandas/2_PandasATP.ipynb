# Pandas - ATP

Evaluamos la información del archivo:

```python
import pandas as pd
import numpy as np
df = pd.read_csv("atp.csv")
print(df.info())
```

```python
#Muestro los primeros 3 registros de la tabla
df.head(3)
```

```python
df[["ATP", "Tournament", "Round", "Date"]].head(3)
```

```python
df.groupby("Tournament").mean()
```

```python
#Convierto los NaN en 0 
filtrado = pd.DataFrame(df)
filtrado = filtrado.replace(np.nan, 0)
filtrado = filtrado.replace("NR", 0)
filtrado.head(3)
```

```python
filtrado.describe(include=[np.number])
```

```python
# Lista de las columnas
list(filtrado)
```

```python
#Seleccionar varios registros salteados
filtrado.iloc[[1,2,4]]
```

```python
#Primeras 3 columnas
filtrado.iloc[:,0:3]
```

```python
#Selección de filas y columas
filtrado.iloc[[1,4,13],[1,4,5]]
```

```python
#Rango de filas y columnas
filtrado.iloc[0:3,2:5]
```

```python
#Defino a Location como index y luego busco todos los índice de Melbourne
filtrado.set_index("Location", inplace=True) #Inplace True para que se lleve a cabo en las siguientes líneas
filtrado.loc["Melbourne"]
```

```python
#Fila y columna
filtrado.loc["Atlanta", "Surface"]
```

```python
filtrado.loc[["Atlanta"],["Surface", "Date"]]
```

```python
#Selección solamente de Grand Slam
filtrado.loc[filtrado["Series"].str.endswith("Slam")]
```

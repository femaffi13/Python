# Pandas - Medical Exam

```python
import pandas as pd
import numpy as np
import seaborn as sns 
df = pd.read_csv('medical_exam.csv')
df.head()
```

Agregar la columna 'overweight': 

```python
df['overweight'] = None
df.head()
```

```python
imc = (df['weight']) / ((df['height']/100)**2)
df['overweight'] = imc
df.head()
```

```python
df0 = df[df['overweight'] >= 25]
df0.overweight = 1
df0.shape
```

```python
df1 = df[df['overweight'] < 25]
df1.overweight = 0
df1.shape
```

```python
df2 = pd.concat([df0,df1])
df2 = df2.sort_index()
df2.head()
```

```python
normalize1 = df2[(df2['cholesterol'] <= 1)]
normalize1 = df2[(df2['gluc'] <= 1)]
normalize1['cholesterol'] = 0
normalize1['gluc'] = 0
normalize1.shape
```

```python
normalize2 = df2[(df2['cholesterol'] > 1)]
normalize2 = df2[(df2['gluc'] > 1)]
normalize2['cholesterol'] = 1
normalize2['gluc'] = 1
normalize2.shape
```

```python
normalize1.shape[0] + normalize2.shape[0]
```

```python
normalize = pd.concat([normalize1, normalize2])
normalize.head(3)
```

```python
dfgrafico = normalize[['cholesterol', 'gluc', 'smoke', 'alco', 'active', 'cardio', 'overweight']]
dfgrafico0 = dfgrafico[(dfgrafico['cardio'] == 0)]
dfgrafico1 = dfgrafico[(dfgrafico['cardio'] == 1)]
print(f'Total: {dfgrafico0.shape[0] + dfgrafico1.shape[0]}')
```

```python
dfgrafico0 = pd.melt(dfgrafico0)
dfgrafico0
```

# Gr√°ficos con Matplotlib

```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

## Barras

```python
x = ['Python','Java','C','C++','Matlab','Octave']
usuarios = [10,7,15,8,9,7]

plt.bar(x,usuarios)
plt.title('Usuarios de Lenguajes de Programaci√≥n')
plt.xlabel('Lenguaje de Programaci√≥n')
plt.ylabel('N√∫mero de usuarios')
plt.show()
```

## Scatter

```python
x = ['F','B','D','J']
y = [1,2,3,4]
colores = ['green', 'red', 'green', 'red']

plt.scatter(x,y, s=150, c=colores, label='texto')
plt.legend(loc=4, frameon=True, title='Leyenda')
plt.show()
```

## Line plots

```python
datos_lins=np.array([1,2,3,4,5,6,7,8]) 
datos_cuads=datos_lins**2

plt.plot(datos_lins,'-o',datos_cuads, '--r')
plt.xlabel('Alg√∫n dato')
plt.ylabel('Alg√∫n otro dato')
plt.title('Un t√≠tulo')
plt.legend(['Dato1','Dato2'])

plt.gca().fill_between(range(len(datos_lins)), datos_lins, datos_cuads, facecolor='green', alpha=0.25)                       
plt.show()
```

NumPy es √∫til para trabajar operaciones con fechas y con Pandas existe una funci√≥n llamada a datetime que permite convertir las fechas NumPy en fechas de biblioteca est√°ndar, que es lo que matplotlib espera. Usamos la funci√≥n map de la biblioteca est√°ndar que devuelve un iterador, matplotlib no puede manejar el iterador, por lo que necesitamos convertirlo en una lista. Para que las fechas no se apilen, podemos obtener un solo eje usando las propiedades del eje x o del eje y del objeto de ejes que podemos obtener, con gca(). Existen muchas propiedades del objeto de ejes, por ejemplo, se puede obtener las l√≠neas de cuadr√≠cula, las ubicaciones de ticks para las etiquetas principales y secundarias, etc. Cada una de las etiquetas es un objeto de texto que a su vez es un artista. Esto significa que se pueden utilizar una serie de funciones de artista diferentes. Una funci√≥n espec√≠fica del texto es la de rotaci√≥n que puede cambiarse en funci√≥n de grados.

```python
datos_lins = np.array([1,2,3,4,5,6,7,8])
datos_cuads = datos_lins**2

plt.figure()
plt.xlabel('Alg√∫n dato')
plt.ylabel('Alg√∫n otro dato')
plt.title('Un t√≠tulo')
plt.legend(['Dato1','Dato2','Dato3'])

datos_observados = np.arange('2021-03-01','2021-03-09',dtype='datetime64[D]')
datos_observados = list(map(pd.to_datetime,datos_observados))

plt.plot(datos_observados,datos_lins,'-o', datos_observados,datos_cuads,'-o')
x=plt.gca().xaxis
for item in x.get_ticklabels():
    item.set_rotation(45)
plt.show()
```

## Bar charts

```python
datos_lins=np.array([1,2,3,4,5,6,7,8])
xvals=range(len(datos_lins))

plt.bar(xvals, datos_lins, width=0.75)
plt.show()
```

```python
datos_lins=np.array([1,2,3,4,5,6,7,8])
datos_cuads=datos_lins**2

xvals=range(len(datos_lins))
plt.bar(xvals, datos_lins, width=0.4)
new_xvals=[]
for item in xvals:
    new_xvals.append(item+0.3)
plt.bar(new_xvals,datos_cuads,width=0.3,color='red')
plt.show()
```

Pueden agregarse barras de error a cada barra, utilizando el par√°metro y error. Por ejemplo, cada uno de nuestros datos, en los datos lineales, podr√≠a ser en realidad un valor medio, calculado a partir de muchas observaciones diferentes. Creamos una lista de valores de error importando una funci√≥n aleatoria:

```python
datos_lins=np.array([1,2,3,4,5,6,7,8])
datos_cuads=datos_lins**2
plt.figure()
xvals=range(len(datos_lins))
plt.bar(xvals,datos_lins,width=0.3)
new_xvals=[]
for item in xvals:
    new_xvals.append(item+0.3)
plt.bar(new_xvals,datos_cuads,width=0.3,color='red')
from random import randint
linear_err=[randint(0,15) for x in range(len(datos_lins))]
plt.bar(xvals,datos_lins,width=0.3,yerr=linear_err)
plt.show()
```

Tambi√©n podemos construir gr√°ficos de barras apilados. Por ejemplo, si quisi√©ramos mostrar valores acumulativos mientras mantenemos la serie independiente, podr√≠amos hacer esto estableciendo el par√°metro inferior y nuestro segundo gr√°fico para que sea igual alprimer conjunto de datos a trazar:

```python
datos_lins=np.array([1,2,3,4,5,6,7,8])
datos_cuads=datos_lins**2
plt.ion()
plt.figure()
xvals=range(len(datos_lins))
plt.bar(xvals,datos_lins,width=0.3)
new_xvals=[]
for item in xvals:
    new_xvals.append(item+0.3)
plt.bar(new_xvals,datos_cuads,width=0.3,color='red')
plt.figure()
xvals=range(len(datos_lins))
plt.bar(xvals,datos_lins,width=0.3,color='b')
plt.bar(xvals,datos_cuads,width=0.3,
bottom=datos_lins,color='r')
plt.show()
```

Finalmente, podemos girar este gr√°fico de barras verticales en un gr√°fico de barras horizontal llamando a la funci√≥n barh, pero teniendo en cuenta que tenemos que cambiar la parte inferior a la izquierda y la altura sobre el ancho:

```python
datos_lins=np.array([1,2,3,4,5,6,7,8])
datos_cuads=datos_lins**2
plt.ion()
plt.figure()
xvals=range(len(datos_lins))
plt.bar(xvals,datos_lins,width=0.3)
new_xvals=[]
for item in xvals:
    new_xvals.append(item+0.3)
plt.bar(new_xvals,datos_cuads,width=0.3,color='red')
plt.figure()
xvals=range(len(datos_lins))
plt.barh(xvals,datos_lins,height=0.3,color='b')
plt.barh(xvals,datos_cuads,height=0.3,left=datos_lins,color='r')
plt.show()
```

## Gr√°ficos circulares

```python
popularity=[56,39,34,34,29]
languages=['Python','SQL','Java','C++','JavaScript']
plt.pie(popularity, labels=languages)
plt.show()
```

```python
popularity=[56,39,34,34,29]
languages=['Python','SQL','Java','C++','JavaScript']
colores=["#13EAC9","#FC5A50","#AAFF32","#FFD700","#9A0EEA"]
plt.pie(popularity, labels=languages, autopct="%0.1f%%", colors=colores)
plt.show()
```

```python
from matplotlib import cm
from matplotlib import colors
popularity=[56,39,34,34,29]
languages=['Python','SQL','Java','C++','JavaScript']
normdata=colors.Normalize(min(popularity),max(popularity))
colormap=cm.get_cmap("GnBu")
colores=colormap(normdata(popularity))
plt.pie(popularity,labels=languages,autopct="%0.1f%%",colors=colores)
plt.show()
```

```python
popularity=[56,39,34,34,29]
languages=['Python','SQL','Java','C++','JavaScript']
colores=["#13EAC9","#FC5A50","#AAFF32","#FFD700","#9A0EEA"]
desfase=(0,0,0,0,0.1)
plt.pie(popularity,labels=languages,autopct="%0.1f%%",colors=colores,explode=desfase)
plt.show()
```

## Diagrama de caja

Tambi√©n se conoce como diagrama de bigotes y se crea para mostrar el resumen del conjunto de valores de datos que tienen propiedades como m√≠nimo, primer cuartil, mediana, tercer cuartil y m√°ximo. En el diagrama de caja, se crea una caja desde el primer cuartil hasta el tercer cuartil, tambi√©n hay una l√≠nea vertical que pasa por la caja en la mediana. Aqu√≠ el eje x denota los datos que se trazar√°n, mientras que el eje y muestra la distribuci√≥n de frecuencia.

```python
jap=np.random.uniform(166,176,100)
ale=np.random.uniform(175,185,100)
arg=np.random.uniform(170,180,100)
plt.boxplot([jap,ale,arg],
notch=True,patch_artist=True,
capprops=dict(color="green"),
medianprops=dict(color="orange"),
whiskerprops=dict(color="red"))
plt.xticks([1,2,3],['Jap√≥n','Alemania','Argentina'])
plt.ylabel('Estaturas(cm)')
plt.show()
```

```python
all_data=[np.random.normal(0,std,100)
for std in range(1,4)]
fig=plt.figure(figsize=(6,4))
bplot=plt.boxplot(all_data,
notch=False,
vert=True,
patch_artist=True)
colors=['pink','lightblue','lightgreen']
for patch,color in zip(bplot['boxes'],colors):
    patch.set_facecolor(color)
plt.xticks([y+1 for y in range(len(all_data))],['x1','x2','x3'])
plt.xlabel('Mediadex')
t=plt.title('Boxplot')
plt.show()
```

# Matplotlib

```python
from matplotlib import pyplot as plt  #pyplot permite generar gráficos.
from matplotlib_venn import venn2  ##Venn2 tiene implementado clases, funciones, métodos, para generar diagramas de Venn.
venn2((1,1,1))  #(izq,der,int)
plt.show()
```

Cada uno de los subconjuntos (3 en diagramas de 2 conjuntos) el módulo utiliza una nomenclatura que consiste en colocar un 1 para indicar que la sección está incluida en el conjunto y un 0 para indicar que está excluida.
De esta manera, siguiendo el orden «ABC», el subconjunto 10(Ab) es el de la izquierda (el que pertenece a Apero no a B); el 01(aB), el de la derecha (el que pertenece a B pero no a A); y el 11(AB), el del medio (la intersección).

```python
venn2({"10":1,"01":2,"11":1})
plt.show()
```

```python
venn2((1,1,1),set_labels=("X","Z")) #Cambia el nombre de los conjuntos
plt.show()
```

```python
venn2((1,1,1),set_colors=("#FF5733","#68FF33")) #Por defecto tiene un alpha (transparencia) de 0.4
plt.show()
```

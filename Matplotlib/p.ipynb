# Matplotlib

```python
from matplotlib import pyplot as plt  #pyplot permite generar gráficos.
from matplotlib_venn import venn2  ##Venn2 tiene implementado clases, funciones, métodos, para generar diagramas de Venn.
venn2((1,1,1))  #(izq,der,int)
plt.show()
```

Cada uno de los subconjuntos (3 en diagramas de 2 conjuntos) el módulo utiliza una nomenclatura que consiste en colocar un 1 para indicar que la sección está incluida en el conjunto y un 0 para indicar que está excluida.
De esta manera, siguiendo el orden «ABC», el subconjunto 10(Ab) es el de la izquierda (el que pertenece a Apero no a B); el 01(aB), el de la derecha (el que pertenece a B pero no a A); y el 11(AB), el del medio (la intersección).

```python
venn2({"10":1,"01":2,"11":1})
plt.show()
```

```python
venn2((1,1,1),set_labels=("X","Z")) #Cambia el nombre de los conjuntos
plt.show()
```

```python
venn2((1,1,1),set_colors=("#FF5733","#68FF33")) #Por defecto tiene un alpha (transparencia) de 0.4
#venn2((1,1,1),set_colors=("#FF5733","#68FF33"),alpha=1) Para quitarle la transparencia
plt.show()
```

La función venn2() retorna una instancia de una clase llamada VennDiagram.  
Los métodos de ésta clase que nos interesan son:  
- get_label_by_id()  
- get_patch_by_id()

```python
diagram = venn2((1,1,1), set_colors=("#FF5733","#68FF33"), alpha=.75)
diagram.get_label_by_id("11").set_text("Intersección") #Modificar texto
diagram.get_label_by_id("11").set_color("#000000") #Label para subconjuntos
diagram.get_patch_by_id("01").set_color("#0000FF") #Patch para conjuntos
diagram.get_patch_by_id("10").set_color("#00FF00")
diagram.get_patch_by_id("11").set_color("#FF0000") #También aplica a subconjuntos
plt.show()
```

Le agregamos líneas con color al diagrama, para eso tenemos que importar el módulo venn2_circles:

```python
from matplotlib_venn import venn2,venn2_circles
diagram=venn2((1,1,1),set_colors=("#FF5733","#68FF33"),alpha=1)
diagram.get_label_by_id("11").set_text("Intersección")
diagram.get_label_by_id("11").set_color("#000000")
diagram.get_patch_by_id("01").set_color("#0000FF")
diagram.get_patch_by_id("10").set_color("#00FF00")
diagram.get_patch_by_id("11").set_color("#FF0000")
venn2_circles(subsets=(1,1,1),color="#000000",alpha=1,linestyle="-.",linewidth=3)
plt.show()
```

Si configuro el círculo del diagrama de Vennen la variable c, puedo llamar a cada círculo individual c [0] y c [1] y establecer el ancho de línea:

```python
diagram=venn2((1,1,1), set_labels=('GrupoA','GrupoB'), set_colors=("#FF5733","#68FF33"), alpha=1)
diagram.get_label_by_id("11").set_text("Intersección")
diagram.get_label_by_id("11").set_color("#000000")
diagram.get_patch_by_id("01").set_color("#0000FF")
diagram.get_patch_by_id("10").set_color("#00FF00")
diagram.get_patch_by_id("11").set_color("#FF0000")
c=venn2_circles(subsets=(1,1,1),color="#000000",alpha=1,linestyle="-.",linewidth=3)
c[0].set_lw(3.0)
c[1].set_lw(5.0)
c[1].set_ls('dotted') #Linea de puntos
plt.title("¡HiceundiagramadeVenn!") #Titulo del diagrama
plt.show()
```

Ahora lo que nos interesa es escribir los valores, estos pueden surgir de una lista, una tupla o un conjunto:

```python
set1 = set(['A','B','C','D'])
set2 = set(['B','C','D','E','F'])
inter = set(set1.intersection(set2))
A = set1-inter
B = set2-inter
diagram = venn2((1,1,1),set_colors=("#FF5733","#68FF33"),alpha=0.5)
diagram.get_label_by_id("11").set_text("Intersección")
diagram.get_label_by_id("11").set_color("#000000")
diagram.get_patch_by_id("01").set_color("#0000FF")
diagram.get_patch_by_id("10").set_color("#00FF00")
diagram.get_patch_by_id("11").set_color("#FF0000")
diagram.get_label_by_id('10').set_text(A)
diagram.get_label_by_id('10').set_fontsize(8)
diagram.get_label_by_id('01').set_text(B)
diagram.get_label_by_id('01').set_fontsize(8)
diagram.get_label_by_id('11').set_text(inter)
diagram.get_label_by_id('11').set_fontsize(12)

c = venn2_circles(subsets=(1,1,1), color="#000000", alpha=0.5, linestyle="dashed", linewidth=3)
c[0].set_lw(3.0)
c[1].set_lw(5.0)
c[1].set_ls('dotted')
plt.show()
```

### Representar mas de 2 conjuntos  
La función venn3() es esencialmente similar. El primer argumento será ahora una tupla de 7 elementos que se corresponden con los siete subconjuntos en el siguiente orden: Abc,aBc,ABc,abC,AbC,aBC,ABC.  
Para esta función venn3 también aplican los argumentos alpha, set_labels y set_colors, como en venn2

```python
from matplotlib_venn import venn3
diagram=venn3((1,1,1,1,1,1,1))
for subset in("111","110","101","100","011","010","001"):
    diagram.get_label_by_id(subset).set_text(subset)
plt.show()
```

Demarcar círculos:

```python
from matplotlib_venn import venn3,venn3_circles
diagram=venn3((1,1,1,1,1,1,1))
for subset in("111","110","101","100","011","010","001"):
    diagram.get_label_by_id(subset).set_text(subset)
venn3_circles(subsets=(1,1,1,1,1,1,1),color="#008000",alpha=1,linestyle="-.",linewidth=3)
plt.show()
```

```python
plt.figure(figsize=(7,5),facecolor='skyblue')
diagram=venn3(subsets=(20,15,4,45,57,23,31),set_labels=('A','B','C'))
for subset in("111","110","101","100","011","010","001"):
    diagram.get_label_by_id(subset).set_fontsize(8)
venn3_circles(subsets=(20,15,4,45,57,23,31),color="#008000",alpha=1,linestyle='dashed',linewidth=2)
plt.title("HiceotrodiagramadeVenn!")
plt.show()
```

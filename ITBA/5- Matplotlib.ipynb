## Histogramas

Para realizar un histograma utilizamos la funci√≥n:

plt.hist(x, bins, propiedades)

x: secuencia de datos
bins: n√∫mero de separaciones de igual ancho
propiedades: opcional
Veamos un ejemplo:

```python
rand = np.random.normal(0,2,1000) # Generamos 1000 numeros al azar con distribuci√≥n normal, con media 0 y desv√≠o estandar 2
plt.hist(rand,20)                 # Creamos un histograma con los numeros al azar, en 20 separaciones
plt.show()
```

Podemos superponer la funci√≥n a la que tiende la distribuci√≥n:

```python
mu = 0              # Media de la distribuci√≥n
sigma = 1           # Desvio estandar
var = sigma**2
t = np.arange(-10,10,0.01)
y = 1/np.sqrt(2*np.pi)*sigma*np.exp(-(t-mu)**2/(2*sigma**2))
rand = np.random.normal(mu,var,10000)
plt.hist(rand,30,density=True)
plt.plot(t,y)
plt.show()
```

## Stem plots

Un stem plot nos permite graficar puntos, se utiliza de la siguiente manera:

plt.stem(x,y)

```python
puntos = 10
espacio = 1

inicio = 0 
fin =  5

x = np.linspace(inicio,fin, puntos)
y = x**2 - 4
y2 = -0.5*x**2-4

plt.subplot(2, 1, 1)
plt.stem(x,y)
plt.subplot(2, 1, 2)
plt.stem(x,y2, 'r-')

plt.show()
```

Al igual que los otros tipos de gr√°ficos, podemos personalizar par√°metros como el formato de linea (linefmt), el formato del punto (markerfmt), el formato de la base (basefmt) y el nombre (label).

```python
puntos = 10
espacio = 1

inicio = 0 
fin =  5

x = np.linspace(inicio,fin, puntos)
y = x**2-4
y2 = -0.5 * x**2 - 4

plt.stem(x, y , linefmt='g--', markerfmt='go', basefmt='k', label='Ingresos')
plt.stem(x, y2, linefmt='r-.', markerfmt='ro', basefmt='k', label = 'Gastos')
plt.plot(x,y+y2,'b-',label = 'Ganancia')

plt.legend()
plt.show()
```

# Scatter Plots

Los scatter plots nos permiten graficar puntos al igual que, por ejemplo, los stem plots. La diferencia radica en la forma en que se representa la informaci√≥n. Suelen ser usados en investigaci√≥n para saber si un conjunto de datos presenta tendencias, o si las variables a comparar est√°n correlacionadas de alguna manera.

```python
N = 50                              # Graficamos N puntos
x = np.random.rand(N)               # Cada eje tiene N valores aleatorios
y = np.random.rand(N)
colors = np.random.rand(N)          # Los colores tambi√©n son aleatorios
area = (30 * np.random.rand(N))**2  # Al igual que el √°rea de las circunferencias

# Graficamos y en funci√≥n de x, con las opciones de forma s, colores c, y transparencia alpha
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
# Mostramos el gr√°fico 
plt.show()
```

Como vemos, si bien la informaci√≥n se representa correctamente, no parece tener mucha utilidad. Ahora veamos lo que pasa cuando existe una cierta correlaci√≥n entre los datos que representamos. En este caso las variables tienen distribuci√≥n normal.

```python
mu = [0, 0]                          # Valores medios de x e y
cxy = [[1, 1], [1, 2]]               # Matriz de covarianza, [[Cxx, Cxy], [Cyx, Cyy]]
x, y = np.random.multivariate_normal(mu, cxy, 5000).T

plt.scatter(x, y, s=5, alpha=0.5)
plt.show()
```

Aqu√≠ podemos ver claramente que las variables est√°n correlacionadas (viendo la forma de elipse que se genera en el gr√°fico). La teor√≠a detr√°s de la correlaci√≥n entre variables aleatorias es muy pesada, as√≠ que no esperamos que se entienda completamente, pero s√≠ el concepto para representar los datos.
